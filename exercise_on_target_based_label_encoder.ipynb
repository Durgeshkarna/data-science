```python
import pandas as pd
import seaborn as sns
```

```python
tips_dataset=sns.load_dataset('tips')
```

```python
tips_dataset.head()
```

```python
tips_dataset.groupby('time')['total_bill']
```

```python
mean_price=tips_dataset.groupby('time')['total_bill'].mean().to_dict()
```

```python
mean_price
```

```python
tips_dataset['time_encoded']=tips_dataset['time'].map(mean_price)
```
